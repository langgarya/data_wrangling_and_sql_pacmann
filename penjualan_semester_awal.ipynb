```python
#Membuat query untuk permasalahan 1
order_cust_join_order_filtered1_query = '''
        select 
            customer_state,
            order_purchase_timestamp
        from 
            olist_order_customer_dataset 
        join 
            olist_order_dataset 
        using(customer_id)
        where strftime('%Y',order_purchase_timestamp)='2018' 
            and (strftime('%m',order_purchase_timestamp) >='01' 
                and strftime('%m',order_purchase_timestamp) <='06')
        '''
order_cust_join_order_filtered1 = pd.read_sql(order_cust_join_order_filtered1_query, con)

#Menampilkan data
counts = order_cust_join_order_filtered1['customer_state'].value_counts()
top5 = counts.nlargest(5)
df_top5 = pd.DataFrame({'customer_state': top5.index, 'Count': top5.values})
sns.barplot(data=df_top5, x='customer_state', y='Count')
plt.show()
```

```python
#Membuat query untuk permasalahan 2
order_cust_join_order_filtered2_query = '''
        select 
            strftime('%m', order_purchase_timestamp) month,
            customer_state,
            count(customer_state) n_customer_state
            
        from 
            olist_order_customer_dataset 
        join 
            olist_order_dataset 
            using(customer_id)
        where 
            customer_state = 'SP' and
            month >= '01' and
            month <= '06' and
            strftime('%Y', order_purchase_timestamp) = '2018'
        group by 1
        '''
order_cust_join_order_filtered2 = pd.read_sql(order_cust_join_order_filtered2_query, con)

#Menampilkan data
sns.lineplot(data=order_cust_join_order_filtered2, 
             x=order_cust_join_order_filtered2['month'], 
             y=order_cust_join_order_filtered2['n_customer_state'], 
             hue=order_cust_join_order_filtered2['customer_state'])
plt.show()
```

```python
#Memeriksa data olist_products_dataset
pd.read_sql("select * from olist_products_dataset", con)
```

```python
#Memeriksa data olist_order_items_dataset
pd.read_sql("select * from olist_order_items_dataset", con)
```

```python
#Menggabungkan data olist_products_dataset dengan olist_order_items_dataset
products_join_orderItems_filtered1_query='''
        select 
            order_id,
            product_category_name
        from
            olist_products_dataset
        join
            olist_order_items_dataset
            using(product_id)
            '''
products_join_orderItems_filtered1 = pd.read_sql(products_join_orderItems_filtered1_query, con)
products_join_orderItems_filtered1
```

```python
#Menggabungkan data olist_order_customer_dataset dengan olist_order_dataset dan melakukan filtering
order_cust_join_order_filtered3_query = '''
        select
            order_id,
            customer_state,
            order_purchase_timestamp
        from
            olist_order_customer_dataset
        join
            olist_order_dataset
            using(customer_id)
        where 
            strftime('%Y''%m''%d', order_purchase_timestamp) <= '2018''06''30' and
            strftime('%Y''%m''%d', order_purchase_timestamp) >= '2018''01''01'
        order by 3 desc
            '''
order_cust_join_order_filtered3 = pd.read_sql(order_cust_join_order_filtered3_query, con)
order_cust_join_order_filtered3
```

```python
#Menggabungkan products_join_orderItems_filtered1_query dengan order_cust_join_order_filtered3_query
join = pd.merge(products_join_orderItems_filtered1, order_cust_join_order_filtered3, on='order_id')
join = join.dropna(subset='product_category_name')


join.groupby(['customer_state', 'product_category_name']).count()
```

```python
#Mengagregasi untuk mendapatkan produk di tiap 5 negara
a=join[(join.customer_state=='SP')]
b=a.groupby(['customer_state', 'product_category_name']).count()\
.sort_values(by='order_id', ascending=False).head(5)

c=join[(join.customer_state=='RJ')]
d=c.groupby(['customer_state', 'product_category_name']).count()\
.sort_values(by='order_id', ascending=False).head(5)

e=join[(join.customer_state=='MG')]
f=e.groupby(['customer_state', 'product_category_name']).count()\
.sort_values(by='order_id', ascending=False).head(5)

g=join[(join.customer_state=='RS')]
h=g.groupby(['customer_state', 'product_category_name']).count()\
.sort_values(by='order_id', ascending=False).head(5)

i=join[(join.customer_state=='PR')]
j=i.groupby(['customer_state', 'product_category_name']).count()\
.sort_values(by='order_id', ascending=False).head(5)
```

```python
df_concat=pd.concat([b,d,f,h,j])
df_concat
```

```python
#Menampilkan data permasalahan 3
sns.barplot(data=df_concat.reset_index(), x='customer_state', y='order_id', hue='product_category_name')
plt.show()
```
